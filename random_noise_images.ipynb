# Random Noise Images

This notebook generates a few simple images made from random noise in base R.

```r
set.seed(123)

make_noise_image <- function(n = 100, palette = gray.colors(256)) {
  noise <- matrix(rnorm(n * n), nrow = n, ncol = n)
  image(
    1:n,
    1:n,
    noise,
    col = palette,
    axes = FALSE,
    xlab = "",
    ylab = "",
    useRaster = TRUE
  )
}
```

```r
par(mfrow = c(1, 3), mar = c(1, 1, 2, 1))

make_noise_image(palette = gray.colors(256))
title("Grayscale Noise")

make_noise_image(palette = heat.colors(256))
title("Heat Noise")

make_noise_image(palette = terrain.colors(256))
title("Terrain Noise")
```
